```python
#Midpoint
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt

#Dumb Function
def f(x):
    return np.sin(np.sqrt(100*x))**2

#Result of integral
intResult = (201 - np.cos(20) - 20*np.sin(20))/400

#Number of Steps
numStep = int(input('Gib steps: '))

#Integration bounds
xmin = 0
xmax = 1

#Binwidth
binWidth = (xmax-xmin)/float(numStep)

# the midpoint rule summation
numericalInt = 0
xi = xmin
for i in range(numStep):
    xii = xi + binWidth
    midPoint = (xi+xii)/2
    numericalInt += f(midPoint)
    xi = xii
numericalInt *= binWidth

print("Numerical value integral: ", numericalInt)
error = np.abs( (numericalInt - intResult) / intResult )
print("Actual value integral:\t  ", intResult)
print("Relative error =\t   {0:.4g}%" .format(error*100.))
print("\n")

# plot function and rectangles
x = np.arange(xmin, xmax, 0.01)
x_mid = np.arange(xmin + binWidth/2, xmax, binWidth)
plt.xlim(xmin, xmax)
plt.plot(x, f(x))
plt.xlabel(r"x")
plt.ylabel(r"f(x)")
plt.bar(x_mid, f(x_mid), align='center', width = binWidth, edgecolor = 'k', alpha=0.5)
plt.title("Midpoint rule", fontsize=16)
plt.show()
```

```python
#Trapt Beatz
from scipy.interpolate import interp1d

numStep = int(input('MORE STEP '))

# trapezoidal rule summation
numericalInt = (f(xmax) + f(xmin))/2
xi = xmin
for i in range(numStep-1):
    xi = xi + binWidth
    numericalInt += f(xi)
numericalInt *= binWidth

# print results
print("Numerical value integral: ", numericalInt)
error = np.abs( (numericalInt - intResult) / intResult )
print("Actual value integral:\t  ", intResult)
print("Relative error =\t   {0:.4g}%" .format(error*100.))
print("\n")

# plot function and trapezoids
x = np.arange(xmin, xmax, 0.01)
x_bound = np.arange(xmin, xmax+binWidth, binWidth)
plt.xlim(xmin, xmax)
plt.plot(x, f(x))
plt.xlabel("x")
plt.ylabel("f(x)")
spline_linear = interp1d(x_bound, f(x_bound), kind='linear')
plt.plot(x, spline_linear(x), 'k')
plt.fill_between(x, 0, spline_linear(x), alpha = 0.4, facecolor = 'b', edgecolor = 'k')
plt.stem(x_bound, f(x_bound), markerfmt = ',')
plt.title("Trapezoidal rule", fontsize=16)
plt.show()
```

```python
#Lisa Simpsons Rule
numberSteps = int(input('Number of steps: '))
binWidth = (xmax - xmin)/float(numberSteps)

# Simpson's rule summation
numericalInt = f(xmax) + f(xmin)
xi = xmin
for i in range(numberSteps-1):
    xi = xi + binWidth
    factor = 2
    if i%2 == 0:
        factor = 4
    numericalInt += f(xi)*factor
numericalInt *= binWidth/3.

# print results
print("Numerical value integral: ", numericalInt)
error = np.abs( (numericalInt - intResult) / intResult )
print("Actual value integral:\t  ", intResult)
print("Relative error =\t   {0:.4g}%" .format(error*100.))
print("\n")

# plot function and parabolas
x = np.arange(xmin, xmax, 0.01)
x_bound = np.arange(xmin, xmax+binWidth, binWidth)
plt.xlim(xmin, xmax)
plt.plot(x, f(x))
plt.xlabel("x")
plt.ylabel("f(x)")
spline_linear = interp1d(x_bound, f(x_bound), 2)
plt.plot(x, spline_linear(x), 'k')
plt.fill_between(x, 0, spline_linear(x), alpha = 0.4, facecolor = 'b', edgecolor = 'k')
plt.stem(x_bound, f(x_bound), markerfmt = ',')
plt.title("Simpson's rule", fontsize=16)
plt.show()
```

```python
#Trapt beatz adaptation
accuracy = float(input('Desired accuracy: '))
numberSteps = 10
factor = 1
error = 1

binWidth = (xmax - xmin)/float(numberSteps)
numericalInt = (f(xmax) + f(xmin))/2
xi = xmin

# first estimate
for i in range(numberSteps-1):
    xi = xi + binWidth
    numericalInt += f(xi)
oldInt = numericalInt*binWidth
    
# improve until desired accuracy
while(error > accuracy):
    xi = xmin + binWidth/2
    for i in range( factor*numberSteps ):
        numericalInt += f(xi)
        xi += binWidth
    binWidth *= 0.5
    factor *= 2
    error = abs(numericalInt*binWidth - oldInt)/3
    oldInt = numericalInt*binWidth
        
numericalInt *= binWidth

print("Accuracy reached using\t  ", numberSteps*factor, " steps")
print("Numerical value integral: ", numericalInt)
print("Estimated numerical error: {0:.4g}%" .format(np.abs(error)*100))
error = np.abs( (numericalInt - intResult) / intResult )
print("Actual value integral:\t  ", intResult)
print("True relative error: \t   {0:.4g}%" .format(error*100))
```
