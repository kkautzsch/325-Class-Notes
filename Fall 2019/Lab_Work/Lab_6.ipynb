```python
#Lab 6
#Relaxation Method
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as opt

x = 1
y = 0

for k in range(10):
    x,y = (2*x**2 - 2*y**3 + 1)/4,(-x**4 - 4*y**4 + 8*y + 4)/12
    print(x)
    print(y)
    
print('\nConverged in',str(k),' iterations', 'x value',x,'y value',y)
```

```python
#Newtons Method
import scipy.optimize as opt


x1 = np.arange(0,1,0.01)
plt.plot(x1, 924*x1**6 - 2772*x1**5 + 3150*x1**4 - 1680*x1**3 + 420*x1**2 -42*x1 + 1, label='Das Function')
plt.plot(x1,0*x1, label='Some line')
plt.xlabel('Some x input')
plt.ylabel('Some Function')
plt.legend()

def P(x):
    return 924*x**6 - 2772*x**5 + 3150*x**4 - 1680*x**3 + 420*x**2 -42*x + 1

def n(x):
    return -P(x)

#Some shid 1
a,b,c = 0.0,0.05,0.2
gold = opt.golden(P,brack=(a,b,c))
print('First Min',gold)

#Some shid 2
a1,b1,c1 = 0.20,0.3,0.4
gold1 = opt.golden(n,brack=(a1,b1,c1))
print('First Max',gold1)
```
