```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import gravitational_constant

mass_sun = 1.9891e30
factor = -mass_sun*gravitational_constant

def f(r, t):
    x=r[0]        #unpack the array r into local variables
    y=r[1]
    X=r[2]        # derivative of x
    Y=r[3]        # derivative of y
    # 4 first order ODEs (transformed from 2 second order ODEs)
    dx = X
    dy = Y
    radius_cubed = (x*x + y*y)**(3/2)
    dX = factor*x/radius_cubed
    dY = factor*y/radius_cubed
    return np.array([dx, dy, dX, dY],float)

# orbit of the Earth
start = 0.0
seconds_per_year = 31536000
num_years = 2
end = num_years*seconds_per_year
numSteps = 10000
stepSize = (end-start)/numSteps

tpoints = np.arange(start,end,stepSize)
xpts = []
ypts = []

# take perihelion as initial condition
perihelion_distance = 1.4710e11
perihelion_velocity = 3.0287e4
# choose axes so that x = 0 at the beginning
# this means that the velocity only has y-component at the beginning
# remember that the velocity vector is tangential to the position vector
r = np.array([0.0,perihelion_distance,perihelion_velocity,0.0],float)   

# all steps are as normal RK4, but vector valued quantities used
for t in tpoints:
    xpts.append(r[0])
    ypts.append(r[1])

    k1 = stepSize*f(r,t)
    k2 = stepSize*f(r+0.5*k1,t+0.5*stepSize)
    k3 = stepSize*f(r+0.5*k2,t+0.5*stepSize)
    k4 = stepSize*f(r+k3,t+stepSize)
    r += (k1+2.*k2+2.*k3+k4)/6

plt.plot(tpoints,xpts,label='x')
plt.plot(tpoints,ypts,label='y')
plt.xlabel("t")
plt.legend(loc='upper right')
plt.show()

plt.plot(xpts,ypts)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2e11,2e11)
plt.ylim(-2e11,2e11)
plt.gca().set_aspect('equal', 'datalim')
plt.show()
```

```python
# same using scipy.integrate

from scipy.integrate import solve_ivp, odeint

# the order of arguments is different
def f_mod(t, r):
    return f(r, t)

initial = [0.0,perihelion_distance,perihelion_velocity,0.0] 

sol = solve_ivp(f_mod, (start, end), initial, method='RK45', rtol = 1e-8)
print(sol.message)

sol2 = odeint(f, initial, tpoints)

plt.plot(sol.y[0], sol.y[1], label='Runge-Kutta with solve_ivp')
plt.plot(sol2[:,0], sol2[:,1], ':', label = 'odeint library')
plt.plot(xpts, ypts, '--', label='Runge-Kutta explicit')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.gca().set_aspect('equal', 'datalim')
plt.show()
```

```python
# same for a comet

perihelion_distance = 4e12
perihelion_velocity = 500
end = 100*seconds_per_year        # one year is longer for a comet

initial = [0.0,perihelion_distance,perihelion_velocity,0.0]  
sol = solve_ivp(f_mod, (start, end), initial, method='RK45', rtol=1e-5)
print(sol.message)

plt.plot(sol.y[0], sol.y[1], label='Runge-Kutta')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.gca().set_aspect('equal', 'datalim')
plt.show()
```

```python
import numpy as np

# import physical constants from SciPy
from scipy.constants import hbar, m_e, e

a = 1e-11
V0 = 50*e
xmin = -5*a
xmax = 5*a
interval = xmax - xmin

# numerical parameters
numSteps = 1000
stepSize = interval/numSteps

# anharmonic potential
def V(x):
    return V0*x**4/a**4

# Schroedinger equation
def f(r,x,E):
    psi = r[0]
    phi = r[1]    # this is what we will call the derivative of psi
    fpsi = phi    # we turned the second order equation into two first order equations, as usual
    fphi = (2*m_e/hbar**2)*(V(x)-E)*psi
    return np.array([fpsi,fphi],float)

# Calculate the wavefunction for a particular energy with RK4, keep track of full solution
def solve(E):
    psi = 0.0
    phi = 1.0     # the initial condition for phi doesn't matter => only affects normalization
    r = np.array([psi,phi],float)
    solution = []

    for x in np.arange(xmin,xmax,stepSize):
        k1 = stepSize*f(r,x,E)
        k2 = stepSize*f(r+0.5*k1,x+0.5*stepSize,E)
        k3 = stepSize*f(r+0.5*k2,x+0.5*stepSize,E)
        k4 = stepSize*f(r+k3,x+stepSize,E)
        r += (k1+2*k2+2*k3+k4)/6
        solution.append(r[0])

    return solution

# ground state energy using the secant method
E1 = 0.0
E2 = e
psi2 = solve(E1)[-1]   # need the last psi value of the full wavefn solution

target = e/1000
while abs(E1-E2)>target:
    psi1, psi2 = psi2, solve(E2)[-1]
    E1, E2 = E2, E2-psi2*(E2-E1)/(psi2-psi1)

print("E =",E2/e,"eV")
```

```python
# systematically search for excited states: based on Garrett's code

energyList = [E2/e]
wavefnList = [solve(E2)]
E_old = 0

for i in np.arange(0, 5000, 200):
    E1 = e*i
    E2 = e*(i+199)
    psi2 = solve(E1)[-1]

    target = e/1000
    while abs(E1 - E2) > target:
        psi1 = psi2
        psi2 = solve(E2)[-1]
        E1, E2 = E2, E2 - psi2 * (E2 - E1) / (psi2 - psi1)

    if int(E2 / e) != int(E_old / e):
        E_old = E2

    if int(E2 / e) != int(energyList[-1]):
        energyList.append(E2 / e)
        wavefnList.append(solve(E2))

    if len(energyList) >= 3:
        break
        
print("Ground State =         ", energyList[0], "eV")
print("First Excited State =  ", energyList[1], "eV")
print("Second Excited State = ", energyList[2], "eV")
```

```python
# plot non-normalized wavefunctions

import matplotlib.pyplot as plt

x = np.arange(xmin, xmax, stepSize)
for i in wavefnList:
    plt.plot(x, i)

plt.show()

# I don't see an issue with very large values
```

```python
# wavefunction normalization
from scipy.integrate import simps

normList = np.sqrt(simps(np.square(wavefnList), x=x))

state0 = wavefnList[0]/normList[0]
state1 = wavefnList[1]/normList[1]
state2 = wavefnList[2]/normList[2]

plt.plot(x, state0, label='ground state')
plt.plot(x, state1, label='1st excited state')
plt.plot(x, state2, label='2nd excited state')
plt.xlim(xmin, xmax)
plt.legend()
plt.xlabel("x [meters]")
plt.ylabel("normalized wavefunctions")
plt.show()
```
